```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from keras.preprocessing import image
from keras.applications.inception_v3 import preprocess_input

base_dir = '/media/baxter/DataDisk/Cornell Grasps Dataset/original'
train_dir = os.path.join(base_dir, 'train')

base_filename = 'pcd0'
filename_suffix = 'png'


for img_instance in range(0, 4):
    img_filename = os.path.join(train_dir, base_filename + str(100+img_instance) + "r" + "." + filename_suffix)
    img = Image.open(img_filename)
    fig = plt.figure()
    plt.subplot(2, 2, img_instance+1)
    plt.imshow(img)
    plt.show()
```
